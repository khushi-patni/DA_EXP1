Import Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline
```

Importing the Dataset

```python
df = pd.read_csv("mtcars.csv")
```

```python
df.head()
```

Describing the Dataset

```python
df.describe()
```

Line Plot

```python
plt.figure(figsize=(16,9)) # Figure size
sns.lineplot(x='model', y='hp', data=df, marker='o', color='blue') 
plt.title('Horse Power') # Title
plt.xticks(df.model.unique(), rotation=45) 
plt.show()
```

Bar Graph

```python
x=df['model'].values
y=df['hp'].values
fig, ax = plt.subplots(figsize =(12, 9))   

ax.barh(x, y, color='blue', edgecolor=(0,0,0)) 
ax.set_title('Horse Power') 
plt.show() 
```

Pie Chart

```python
my_labels=df['model']
plot = df.plot.pie(y='hp', figsize=(15, 15), labels=my_labels, autopct='%1.1f%%', subplots=True)
```

Correlation of the Dataset

```python
sns.set_theme()
corrmat = df.iloc[:,1:].corr()
ax = sns.heatmap(corrmat)
ax
```
